```python
from kaia.eaglesong.core import *
from kaia.eaglesong.drivers.ipython import *

def main(context: BotContext):
    yield f'Hello. Your user_id is {context.user_id}. Say anything and I will repeat. Or /start to reset.'
    while True:
        yield Listen()
        yield context.input

        
root = Routine.interpretable(main, PushdownFilter)
interpreter = IPythonInterpreter(root, IPythonChatModel())
chat = IPythonChatWidget(interpreter)
chat.run()
```
